```python
import re

#setting up punctuation
replace_n_space = re.compile("(\.)|(\;)|(\:)|(\!)|(\')|(\?)|(\,)|(\")|(\|)|(\()|(\))|(\[)|(\])|(\%)|(\$)|(\>)|(\<)|(\{)|(\})")
replace_w_space = re.compile("(<br\s/><br\s/?)|(-)|(/)|(:).")
```

```python
import preprocessor as p

def clean_tweets(df):
  tempArr = []
  for line in df:
    tmpL = p.clean(line)
    tmpL = replace_n_space.sub("", tmpL.lower())
    tmpL = replace_w_space.sub(" ", tmpL)
    tempArr.append(tmpL)
  return tempArr
```

```python
train_tweet = clean_tweets(train["tweet"])
train_tweet = pd.DataFrame(train_tweet)
```
